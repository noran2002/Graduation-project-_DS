```python
df = pd.read_csv(os.path.join('data', 'food', 'Translated_dataset.csv'))
df.columns = df.columns.str.strip()
df.columns = df.columns.str.capitalize()
df
```

```python
df.reset_index(drop=True)
```

```python
df.columns = df.columns.str.strip()
df['Diabeties'] = df['Disease'].str.contains('diabeties')
df['Heart_Disease'] = df['Disease'].str.contains('heart_disease')
df['Hypertension'] = df['Disease'].str.contains('hypertension')
df
```

```python
from sklearn.preprocessing import LabelEncoder

# Sample categorical data

# Create an instance of LabelEncoder
encoder = LabelEncoder()

# Fit and transform the data
df['Diabeties'] = encoder.fit_transform(df['Diabeties'])
df['Heart_Disease'] = encoder.fit_transform(df['Heart_Disease'])
df['Hypertension'] = encoder.fit_transform(df['Hypertension'])
df
```

```python
df.drop('Disease', inplace=True, axis=1)
df.drop('Price', inplace=True, axis=1)
df = df.apply(lambda x: x.strip() if isinstance(x, str) else x)
df
```

```python
df.to_csv(os.path.join('data', 'food', 'Food_Data.csv'))
df=pd.read_csv(os.path.join('data', 'food', 'Food_Data.csv'))
df.reset_index()
df = df.drop(columns=['Unnamed: 0'])
df = df.drop(columns=['Unnamed: 1'])
df.to_csv(os.path.join('data', 'food', 'Food_Data.csv'))
df
```
